**Data Normalization** #to convert values from 0 to 1

```python
mydata[['width', 'length', 'height']].head(20)
```

```python
#simple feature scaling

mydata['width'] = mydata['width']/mydata['width'].max()

mydata['length'] = mydata['length']/mydata['length'].max()

mydata['height'] = mydata['height']/mydata['height'].max()

mydata[['width', 'length', 'height']].head(20)
```

**Get Dummies**

```python
dummies_fueltype =pd.get_dummies(mydata['fuel-type'])

dummies_fueltype.head(20)

#created variables to convert the fuel type column to two columns (gas/diesel)
```

```python
dummies_fueltype.rename(columns = {'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'},inplace =True)
                                   
dummies_fueltype.head()
```

```python
df = pd.concat([mydata, dummies_fueltype], axis = 1)

df.head(10)

#combine original dataframe with the dummies
```

```python
df.shape
```

```python
dummies_aspiration = pd.get_dummies(mydata['aspiration']) # this function is to convert the aspiration column into two columns

dummies_aspiration.head(20)

```

```python
dummies_aspiration.rename(columns = {'std':'aspiration-type-std', 'turbo': 'aspiration-type-turbo'}, inplace = True)

dummies_aspiration.head()
```

```python
ef = pd.concat([df, dummies_aspiration], axis = 1) # this function is used to add/concetenate the two additional columns to the previous data set

ef.head(10)
```
